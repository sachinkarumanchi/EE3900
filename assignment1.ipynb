```python
import numpy as np
import matplotlib.pyplot as plt

#entering input points
P=np.array([4,3]).reshape((2,1))
Q=np.array([-2,2]).reshape((2,1))
#PQ is the distance between points P and Q calculating using norm
PQ=P-Q
d=np.linalg.norm(PQ)
print('The distance between two points P and Q is',d)
#for plotting the points on the graph reconstructing arrays using P and Q
X=np.array([4,-2])
Y=np.array([3,2])
#plotting the points on graph
plt.scatter(X,Y,label='points',color='orange')
plt.plot(X,Y,label='PQ')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Distance between two points')
plt.annotate("P:[4,3]", (4,3)) 
plt.annotate("Q:[-2,2]",(-2,2))
plt.legend()
plt.grid(linestyle='--')
plt.axis([-3,5,1.8,3.2])
plt.show()
```
