```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA

def line_gen(A,B):
  len =10
  dim = A.shape[0]
  x_AB = np.zeros((dim,len))
  lam_1 = np.linspace(0,1,len)
  for i in range(len):
    temp1 = A + lam_1[i]*(B-A)
    x_AB[:,i]= temp1.T
  return x_AB


def circ_gen(C,O,r):
	len = 50
	theta = np.linspace(0,2*np.pi,len)
	x_circ = np.zeros((2,len))
	x_circ[0,:] =C[0,:] + r*np.cos(theta)
	x_circ[1,:] = C[1,:]+ r*np.sin(theta)
	x_circ = (x_circ.T + O).T
	return x_circ

def line_dir_pt(m,A,k1,k2):
  len = 10
  dim = A.shape[0]
  x_AB = np.zeros((dim,len))
  lam_1 = np.linspace(k1,k2,len)
  for i in range(len):
    temp1 = A + lam_1[i]*m
    x_AB[:,i]= temp1.T
  return x_AB

#centre and radius of circles
A=np.array([0,0])
r1=6
r2=4
#centre and radius of circles
B=np.array([0,0])
r1=6
r2=4

#Input parameters
A=np.array(([0,0]))
B=np.array(([8,0]))
P=np.array(([2,3.4641016151377544]))
Q=np.array(([2,-3.4641016151377544]))
R=np.array(([6.875,2.7810744326608736]))
S=np.array(([6.875,-2.7810744326608736]))

f = -16
m = np.array(([1,-0.5773502691896258]))
n = np.array(([1,0.5773502691896258]))
M = np.array(([1,-0.40451991747794525]))
N = np.array(([1,0.40451991747794525]))
k1 = 10
k2 = -10
##Generating all lines
x_BP = line_dir_pt(m,B,k1,k2)
x_BQ = line_dir_pt(n,B,k1,k2)
x_AB = line_gen(A,B)
x_AP = line_gen(A,P)
x_AQ = line_gen(A,Q)
x_AR = line_dir_pt(M,A,k1,k2)
x_AS = line_dir_pt(N,A,k1,k2)
x_BR = line_gen(B,R)
x_BS = line_gen(B,S)


#Plotting all lines
plt.plot(x_AB[0,:],x_AB[1,:],label='$AB$')
plt.plot(x_AP[0,:],x_AP[1,:],label='$AP$')
plt.plot(x_BP[0,:],x_BP[1,:],label='$BP$')
plt.plot(x_BQ[0,:],x_BQ[1,:],label='$BQ$')
plt.plot(x_AQ[0,:],x_AQ[1,:],label='$AQ$')
plt.plot(x_BR[0,:],x_BR[1,:],label='$BR$')
plt.plot(x_AR[0,:],x_AR[1,:],label='$AR$')
plt.plot(x_BS[0,:],x_BS[1,:],label='$BS$')
plt.plot(x_AS[0,:],x_AS[1,:],label='$AS$')



plt.plot(A[0], A[1], 'o')
plt.text(-0.9,-0.9,'A',weight = "bold")
plt.plot(B[0], B[1], 'o')
plt.text(8,0.5,'B',weight = "bold")
plt.plot(P[0], P[1], 'o')
plt.text(2,3.8,'P', weight = "bold")
plt.plot(Q[0], Q[1], 'o')
plt.text(2,-3.8,'Q', weight = "bold")
plt.plot(R[0],R[1],'o')
plt.text(7,3,'R',weight='bold')
plt.plot(S[0],S[1],'o')
plt.text(7,-3,'S',weight='bold')

tri_coords = np.vstack((Q,R,A)).T
plt.scatter(tri_coords[0,:], tri_coords[1,:])
vert_labels = ['Q','R','A']



#Plotting the first circle of radius 4
c1 = 4
theta1 = np.linspace(0,2*np.pi,50)
x1 = c1*np.cos(theta1)
y1 = c1*np.sin(theta1)

plt.plot(x1,y1,label='$circle(r=4)$')

#Plotting the second circle of radius 3
c2 = 3
theta2 = np.linspace(0,2*np.pi,50)
x2 = 8+c2*np.cos(theta2)
y2 = c2*np.sin(theta2)

plt.plot(x2,y2,label='$circle(r=3)$')


plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.grid() # minor
plt.axis([-6,20,-9,9])
plt.savefig('Assignment3.jpg')
plt.show()
```
