```python
import numpy as np
import matplotlib.pyplot as plt

#for plotting discrete-time signal h[n],g[n]
#h_n,g_n are the outputs of 
n=np.linspace(-6,6,13)
h_n=np.array([0,0,0,-1,-2,-3,4,3,2,1,0,0,0])
g_n=np.array([0,0,0,0,0,-3,4,3,0,0,0,0,0])

#plotting discrete-time signal h[n]
plt.stem(n,h_n)
plt.xlabel('n')
plt.ylabel('$h[n]$')
plt.title('Discrete-time signal $h[n]$')
plt.grid(alpha=1,linestyle='--')
plt.show()

#plotting discrete time signal g[n] 
plt.stem(n,g_n)
plt.xlabel('n')
plt.ylabel('$g[n]$')
plt.title('Discrete-time signal $g[n]$')
plt.grid(alpha=1,linestyle='--')
plt.show()
```
