```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-5,5,0.5)
y1=3*x+1
y2=(x+3)/2
y3=((2*np.sqrt(2)-1)/(3-np.sqrt(2)))*x+4

plt.plot(x,y1,label='0.0.1')
plt.plot(x,y2,label='0.0.2')
plt.plot(x,y3,label='0.0.3')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.axis([-5,5,-3,8])
plt.legend()
plt.show()
#(0.0.1),(0.0.2),(0.0.3) are the lines from tex file
```
